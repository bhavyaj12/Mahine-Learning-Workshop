```python
# Import Libraries
import numpy as np
```

```python
# Load Text For Training
def load_text(filename):
    with open(filename,encoding="utf-8") as f:
        return f.read().lower()
```

```python
data=load_text("data.txt")
```

```python
# Prepare Markov chain transition table
def CreateTransitionTable(data,k=4):
    T={}
    for i in range(len(data)-k):
        X=data[i:i+k]
        Y=data[i+k]
        
        if T.get(X) is None:
            T[X]={}
            T[X][Y]=1
        elif T[X].get(Y) is None:
            T[X][Y]=1
        else:
            T[X][Y]+=1
    return T
```

```python
T=CreateTransitionTable(data)
T
```

```python
def ctp(T):
    for kx in T.keys():
        factor=(sum(T[kx].values()))
        for val in T.get(kx).keys():
            T[kx][val]=T[kx][val]/factor
    return T
            
```

```python
T=ctp(T)
T
```

```python
# Write prediction function
def predict(ctx,T,k=4):
    if T.get(ctx) is None:
        return " "
    sample_arr=list(T[ctx].keys())
    sample_probab=list(T[ctx].values())
    #print(sample_arr,sample_probab)
    return np.random.choice(sample_arr,p=sample_probab)
```

```python
# Sampling
# Generate 
sentence="कौन"
k=4
for i in range(100):
    char_returned=predict(sentence[-k:],T)
    #print(sentence[-k:])
    sentence+=char_returned
print(sentence)
```
